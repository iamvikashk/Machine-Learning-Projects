```python
import numpy as np
import math
x = np.array([92, 56, 88, 70, 80, 49, 65, 35, 66, 67])
y = np.array([98, 68, 81, 80, 83, 52, 66, 30, 68, 73])
```

```python
def gradient_descent():
    n = len(x)
    w = b = 0
    iterations = 100
    learning_rate = 0.0002
    for i in range(iterations):
        y_pred = w * x + b
        cost = (1/n) * sum([val**2 for val in (y - y_pred)])
        wd = -(2/n) * sum(x * (y - y_pred))
        bd = -(2/n) * sum(y - y_pred)
        w = w - learning_rate * wd
        b = b - learning_rate * bd
        cost1 = (1/n) * sum([val**2 for val in (y - (w * x + b))])
        if math.isclose(cost,cost1,rel_tol=1e-5):
            break
    print("w {} b {} cost {} iter {} cost1 {}".format(w,b,cost,i,cost1))
gradient_descent()
```
